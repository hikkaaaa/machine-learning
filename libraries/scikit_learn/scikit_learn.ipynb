# **One Hot Encoder**

```python
d = {'sales': [100000, 222000, 1000000, 522000, 11111, 222222, 75000, 90000, 5222200],
     'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonvile', 'Orlando', 'Jacksonvile', 'Tampa', 'Orlando', 'Tampa'],
     'size': ['Small', 'Medium', 'Large', 'Small', 'Large', 'Large', 'Medium', 'Small', 'Medium']}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df['city'].unique()
```

```python
#one hot encoder is gonna expand the table with all the city unique values
#(we chose 'city', because it doesnt contain numeric values and unlike size,
#we are not able to count it)
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(handle_unknown = 'ignore', sparse_output = False).set_output(transform = 'pandas')
```

```python
ohetransform = ohe.fit_transform(df[['city']])
```

```python
ohetransform
```

```python
df.head()
```

```python
ohetransform.head()
```

```python
df = pd.concat([df, ohetransform], axis = 1).drop(columns = ['city'])
```

```python
df
```

# **Ordial Encoder with Python ML**

```python
d = {'sales': [100000, 222000, 1000000, 522000, 11111, 222222, 75000, 90000, 5222200],
     'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonvile', 'Orlando', 'Jacksonvile', 'Tampa', 'Orlando', 'Tampa'],
     'size': ['Small', 'Medium', 'Large', 'Small', 'Large', 'Large', 'Medium', 'Small', 'Medium']}
```

```python
df = pd.DataFrame(d)
```

```python
df.head()
```

```python
#we gonna use size, since there is no specific order
df['size'].unique()
```

```python
sizes = ['Small', 'Medium', 'Large']
```

```python
from sklearn.preprocessing import OrdinalEncoder
```

```python
enc = OrdinalEncoder(categories = [sizes])
```

```python
enc.fit_transform(df[['size']])
```

```python
df['size'] = enc.fit_transform(df[['size']])
```

```python
df.head()
```

```python
#0 - small,1 - medium, 2 - large
```

# **Handling Missing Data with simple imputer**

```python
import numpy as np
```

```python
miles = pd.DataFrame({'farthest_run_mi': [50.0, 62.0, np.nan, 100.0, 26.0, 13.0, 31.0, 50.0]})
```

```python
miles
```

```python
miles.isna()
```

```python
miles.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
#filling missing values with eman
imp_mean = SimpleImputer(strategy = 'mean')
```

```python
imp_mean.fit_transform(miles)
```

```python
#filling missing values with median, the most common one
imp_median = SimpleImputer(strategy = 'median')
```

```python
imp_median.fit_transform(miles)
```

```python
#filling missing values with the most common value in the dataset
imp_mode = SimpleImputer(strategy = 'most_frequent')
```

```python
imp_mode.fit_transform(miles)
```

```python
#filling missing vlaues with the constant value, the value is 13 in this case
imp_constant = SimpleImputer(strategy = 'constant', fill_value = 13)
```

```python
imp_constant.fit_transform(miles)
```

```python
#filling the missing value in the categorical data
names = pd.DataFrame({'names': ['Jackson', 'Nowlan', 'Honus', 'Some', np.nan, 'Ruth']})
```

```python
names
```

```python
imp_constant_cat = SimpleImputer(strategy = 'constant', fill_value = 'babe')
```

```python
imp_constant_cat.fit_transform(names)
```

```python
#gonna buils a brand new column with 0s and 1s, 1 - nan value
imp_mean_marked = SimpleImputer(strategy = 'mean', add_indicator = True)
```

```python
imp_mean_marked.fit_transform(miles)
```

```python
#more advanced dataset version
#(changing th missing values in both categorical and numerical at the same time)
df = pd.DataFrame({
    'Name': ['Ryan', 'Nolan', 'Walter', 'Honus', 'Christy', np.nan, 'Napoleon', 'Tris'],
    'farthest_run_mi': [50.0, 62.0, np.nan, 100.0, 26.0, 13.0, 31.0, 50.0]
})
```

```python
df
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (imp_constant_cat, ['Name']),
    (imp_mean, ['farthest_run_mi']),
    remainder = 'drop'
)
```

```python
ct.set_output(transform = 'pandas')
```

```python
df_pandas = ct.fit_transform(df)
```

```python
df_pandas
```

# **Simplify Data Preprocessing with Python's Column Transformer**

```python
d = {'sales': [100000, 222000, 1000000, 522000, 11111, 222222, 75000, 90000, 5222200],
     'city': ['Tampa', 'Tampa', 'Orlando', 'Jacksonvile', 'Orlando', 'Jacksonvile', 'Tampa', 'Orlando', 'Tampa'],
     'size': ['Small', 'Medium', 'Large', 'Small', 'Large', 'Large', 'Medium', 'Small', 'Medium']}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
#transforming city with one hot encoder and size with ordinal encoder
from sklearn.preprocessing import OneHotEncoder, OrdinalEncoder
```

```python
ohe = OneHotEncoder(sparse_output = False)
```

```python
ode = OrdinalEncoder()
```

```python
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (ohe, ['city']),
    (ode, ['size']),
    remainder = 'passthrough'
)
```

```python
ct.set_output(transform = 'pandas')
```

```python
df_pandas = ct.fit_transform(df)
```

```python
df_pandas
```
