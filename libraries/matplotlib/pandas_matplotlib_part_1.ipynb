# **Histograms**

```python
import pandas as pd
import numpy as np
%matplotlib inline
from numpy.random import randn
#matplotlib - charting graphs, graphical stuff with pandas

#create a quich datafarame
df = pd.DataFrame(randn(100, 4), columns = ["Mon", "Tues", "Wed", "Thur"])#df with 100 rows and 4 columns with random number data
                                                                          #names of the columns are given
df.sample(5)
```

```python
#histogram
df['Wed'].hist()
```

```python
#histogram with smaller bins, 50, 20 eg
df['Wed'].hist(bins = 50)
```

```python
df['Wed'].hist(bins = 20)
```

```python
#another way without grid bars
df['Wed'].plot(kind = 'hist', bins = 100)
```

```python
#you can seet the parameters of hist and do all sorts of thing
#eg getting rid of the grid and setting a legend to th histogram
df['Wed'].hist(bins = 50, legend = True, grid = False)
```

# **Area Plots with Mathplotlib**

```python
#area plot
df.plot(kind = 'area', stacked = False)
#without stacked = False, it wouldnt work, bcus in our data we have both positive and negative values
#however, it should have only either of those with the default value stacked = True
#because stacked=False means: Each column in the DataFrame is plotted as a separate area, but not stacked on top of each other.
```

```python
#looks quite messy, so lets take the aboulte value
df.abs().plot(kind = 'area', stacked = False)
```

```python
#shading
df.abs().plot(kind = 'area', alpha = 0.3)
#alpha can take values between 0 and 1
#alpha can make the graph transparent
```

```python
#setting the title
df.abs().plot(kind = 'area', alpha = 0.3, title = 'Something crazy')
```

```python
#legend is set as default, we can get rid of it
df.abs().plot(kind = 'area', alpha = 0.3, title = 'Something crazy', legend = False)
```

```python
#plot only on column
df['Mon'].abs().plot(kind = 'area', alpha = 0.3, title = 'something crazy', legend = True)
```

```python
#plot 2 columns
df[['Mon', 'Tues']].abs().plot(kind = 'area', alpha = 0.5, title = 'etcetc', legend = True, stacked=False)
#and you can add as many columns as you want into the list
```

```python
#add table
df.abs().plot(kind = 'area', stacked = 'False', title = 'something crazy', table = True)
```

# **Bar Charts with Matplotlib**

```python
df = df.head(25)
df
```

```python
#bar plot
df.plot(kind = 'bar')
```

```python
#to make it look better, let'd convert everything to absolute values
df.abs().plot(kind = 'bar')
```

```python
#bar plot stacked
df.abs().plot(kind = 'bar', stacked = True)
#stacked - being stacked one up to the other, stacked = False - is the default value
```

```python
#adding grid lines
df.abs().plot(kind = 'bar', grid = True, stacked = True)
```

```python
#shading - transparancy
df.abs().plot(kind = 'bar', stacked = True, alpha = 0.5, grid = True)
```

```python
#title
df.abs().plot(kind = 'bar', stacked = True, alpha = 0.5, grid = True, title = 'something crazy')
```

```python
#to get rid of legend
df.abs().plot(kind = 'bar', alpha = 0.5, title= ' something crazy', stacked = True, legend = False)
```

```python
#another way of declaring and working with bars:
df.abs().plot.bar()
#you can do all the parameters like title, stacked, and etc in the bar() as well
```

# **Line Charts with Matplotlib**

```python
df.plot(kind = 'line')
```

```python
#taking only positive
df.abs().plot(kind = 'line')
```

```python
#line width
df.abs().plot(kind = 'line', lw = 5)
```

```python
#single line
df['Mon'].abs().plot(kind = 'line', lw = 1)
```

```python
#more than one line
df[['Mon', 'Tues']].abs().plot(kind = 'line', lw = 1)
```

```python
#size
df.abs().plot(kind = 'line', lw = 1, figsize = (10, 5))#10 by 5 dimensions
```

```python
#title
df.abs().plot(kind = 'line', title = 'something crazy', lw = 1)
```

```python
#shading - transparancy
df.abs().plot(kind = 'line',lw = 1, alpha = 0.5, title = 'something crazy')
```

```python
#legend
df.abs().plot(kind = 'line', lw = 1, alpha = 0.5, title = 'something crazy', legend = True)
```

```python
#second way line chart
df.abs().plot.line()
```
