# **Scatterplots with Matplotlib**

```python
import pandas as pd
import numpy as np
%matplotlib inline
from numpy.random import randn

df = pd.DataFrame(randn(25, 4), columns = ['Mon', 'Tues', 'Wed', 'Thur'])
df
```

```python
#scatter plot
df.plot(kind = 'scatter', x = 'Mon', y = 'Tues')
#without x and y coordinates, it will generate an error
```

```python
#we can seta  third variable as well
df.plot(kind = 'scatter', x = 'Mon', y = 'Tues', c = 'Wed')
```

```python
#colors
df.plot(kind = 'scatter', x = 'Mon', y = 'Tues', c = 'Wed', cmap = 'magma')
```

```python
#size of a particular column
df.abs().plot(kind = 'scatter', x = 'Mon', y = 'Tues', s = df['Wed'] * 100, alpha = 0.5) #cmap doesnt work here btw
```

```python
#you can also give it a title, legend and etc (they were entioned in the previous charts)
```

# **Box Plots with Matplotlib**

```python
#boxplots
df.plot(kind = 'box')
```

```python
#boxplots size
df.plot(kind = 'box', figsize = (10, 5))
```

```python
#boxplots color
df.plot(kind = 'box', cmap = 'winter')
```

```python
#no legend is available
#but we can do title
df.plot(kind = 'box', cmap = 'winter', title = 'something crazy')
```

```python
#other way
df.plot.box()
```

```python
#adding grid
df.plot(kind = 'box', grid = True)
```

# **Hex Bin Plots with Matplotlib**

```python
#hexplots
df.plot(kind = 'hexbin', x = 'Mon', y = 'Tues')
```

```python
#setting the third variable
df.plot(kind = 'hexbin', x = 'Mon', y = 'Tues', C = 'Wed')
```

```python
#hexbox gridsize
df.plot(kind = 'hexbin', x = 'Mon', y= 'Tues', gridsize = 8)
```

```python
#hexplots colors
df.plot(kind = 'hexbin', x = 'Mon', y = 'Tues', gridsize = 8, cmap = 'BuPu')
```

```python
#title
df.plot(kind = 'hexbin', x = 'Mon', y = 'Tues', C = 'Wed', gridsize = 8, cmap = 'BuPu', title = 'something crazy')
```

```python
#grid
df.plot(kind = 'hexbin', x = 'Mon', y = 'Tues', C = 'Wed', gridsize = 8, cmap = 'BuPu', grid = True)
```

```python
#other method
df.plot.hexbin(x = 'Mon', y = 'Tues')
#setting x and y oordinates are quite necessary
```

# **Density and KDE Plots with Matplotlib**

```python
#density and kernel density estimation plots with scipy
#in case if you are working with jupyter, u need to install a new library - scipy
df.plot(kind = 'kde')
```

```python
#KDE second way
df.plot.kde()
```

```python
#density plot
df.plot(kind = 'density')
```

```python
#title
df.plot(kind = 'density', title = 'something crazy')
```

```python
#changing the transparancy and line weight
df.plot(kind = 'kde', title = ' something crazy', lw = 5, alpha = 0.5)
```

```python
#grid
df.plot(kind = 'kde', title = ' something crazy', lw = 5, alpha = 0.5, grid = True)
```
