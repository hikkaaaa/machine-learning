# **Count data from your pandas DataFrame**

```python
#the DataFrame that we are dealing with
my_df2
```

```python
#count distinct values - descending
my_df2['Color'].value_counts()
```

```python
#count distinct values - ascending
my_df2['Color'].value_counts(ascending = True)
```

```python
#Dog Names - without NaN (if we had an empty data, it would be NaN and it wouldn't show those)
my_df2['DogName'].value_counts()
```

```python
#Dog Names - with NaN (if we had an empty data, it would be NaN and it would still be shown with the rest of the data)
my_df2['DogName'].value_counts(dropna = False)
```

```python
#Get Relative frequency(%)
my_df2['Color'].value_counts(normalize = True)
```

```python
# get specific item count
my_df2['Color'].value_counts()['WHITE']
```

```python
#get a count of all columns across all columns
my_df2.select_dtypes(include='object').apply(pd.value_counts)
```

# **Add new columns to DataFrame**

```python
#the data that we are going to be working with
my_df2
```

```python
#we need a more concise version, so it would be better to shrink the data frame
my_df3 = my_df2.head(6)
my_df3
```

```python
#add column from list, the number of values shoud correspond to the number of rows
gender = ["Male", "Female","Male", "Female", "Female", "Male"]
my_df3["Gender"] = gender
my_df3
```

```python
#add default values - if we don't know the number of rows or the number of rows is too large
my_df3["Alive/Dead"] = [True] * len(my_df3)#will take the whole lenght of the df and add true
my_df3

```

```python
#add NaN values - we can add such values when we are not sure what value we want to assign it to
my_df3["Show Dog"] = [np.nan] * len(my_df3)
my_df3
```

```python
#add column with .insert() - allows position
my_df3.insert(1, "Adopted", [True] * len(my_df3), True) #1th column - right after breeed, we are adding "Adopted" called column, and assigning all the rows of this column to true, True - duplicates are allowed
my_df3
```

```python
#add column with .assign() - created new df
my_df4 = my_df3.assign(Horse = [False] * len(my_df3))#creating a new dataframe and assigning everything to it from the other data frame
#also adding a new column "Horse" and assigning every row to False
my_df4
```

# **Conditional Selections**

```python
#conditional selection - how you filter rows based on certain conditions
#the data we are going to be working on:
my_df3
```

```python
#conditional selection
# > < = >= <= etc
#return booleans
my_df3 == "BROWN" # the points that dont contain "BROWN" are false, and the points that do are true
```

```python
#returns dataframe with data
my_df3[my_df3 == "BROWN"] #only the points that contain "BROWN" will be explicitly shown the BROWN, but those thst do not will have NaN in their boxes
```

```python
#run them on a column
my_df3[my_df3 == "BROWN"]["Color"] #shows the same thing as the previous one, but only the column "Color"
```

```python
#return multiple columns
my_df3[my_df3 == "BROWN"][['Color', 'Breed']]
```
