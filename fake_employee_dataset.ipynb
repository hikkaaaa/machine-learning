```python
#this little assignment will be about a fake employee database and some analysis of it
import numpy as np
import pandas as pd

#create a fake database
df = pd.DataFrame({
    'Name': ['John', 'Jane', 'Bob', 'Alice', 'Alice'],
    'Age': [30, 25, 40, 35, 50],
    'Department': ['police', 'office', 'chief', 'teacher', 'teacher'],
    'Salary': ['500$', '600$', '200$', '582$', '562$'],
    'HireDate': ['07/06/25', '28/05/56', '03/03/25', '30/01/32', np.nan],
    'City': ['Washington', 'San-Francisko', 'China', 'Poland', 'Washington']
})
np.random.seed(0)
df
```

```python
df['Salary'].dtypes #we have it as object, so we have to convert it into string
```

```python
df.dtypes
```

```python
df = df.convert_dtypes()
df.dtypes
```

```python
#converting salary into numeric
df['Salary'] = df['Salary'].str.replace('$', '', regex = False) #we need regex = False bcus of $ sign - since it can represent end of string
                                                                #we are saying that it is just a string, not any special string with regex = False
df['Salary'] = pd.to_numeric(df['Salary'], errors = 'coerce')
df['Salary']
```

```python
#mean salary per department
df.groupby('Department')['Salary'].mean()
```

```python
#count of employees per city
df['City'].value_counts()
```

```python
#department with the highest paid employee
max_salary = df['Salary'].max()
highest_paid_row = df[df['Salary'] == max_salary]
dep = highest_paid_row['Department'].values[0]
name = highest_paid_row['Name'].values[0]
f"Department with the highest paid employee: {name} in {dep}"
```
