```python
import numpy as np
import pandas as pd

df = pd.DataFrame({
    'Owner': ['John', 'Halsey', 'Patricia', 'Oppla', np.nan, 'Di', 'Welsey', np.nan, 'Prov', 'Di', 'Welsey', np.nan, 'Prov'],
    'DogName': ['Katley', 'Welsey', 'Kaui', 'Kaui', 'Husk', 'Husk', 'Husk', 'Pola', 'Pola', np.nan, 'ola', 'Welsey', 'Kaui'],
    'ZIPCode': [123, 125, 456, 586, 256, 563, 785, 58636, 2655, 254, 258, 2533, 215]
})
np.random.seed(0)
df
```

```python
#first lets try to get rid of nan values, i want to change them to 0
df.isnull().sum()
```

```python
df[['Owner', 'DogName']] = df[['Owner', 'DogName']].fillna('0')
df
```

```python
#not lets check the datatype, and convert everything to a compatible one just for the future
df.dtypes
```

```python
df = df.convert_dtypes()
df.dtypes
```

```python
#okay now lets find the amount of unique DogNames
df['DogName'].value_counts()
```

```python
df['DogName'].unique()
```

```python
f"The dog that appears the most is: {df['DogName'].value_counts().index[0]}"
```

```python
#average owner zip code
df['ZIPCode'].mean()
```

```python
#add a new column
df['IsPopular'] = ['True' if x > 500 else 'False' for x in df['ZIPCode']]
df
```
