# **My First ML Project**

# **Load Data**

```python
import pandas as pd
```

```python
df = pd.read_csv('https://raw.githubusercontent.com/dataprofessor/data/refs/heads/master/delaney_solubility_with_descriptors.csv')
```

```python
df.head()
```

# **Data Preparation**

### Data seperation as X and y

```python
y = df['logS']
```

```python
y.head()
```

```python
X = df.drop('logS', axis = 1)
```

```python
X.head()
```

### Data Splitting

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 100)
```

```python
X_train #x-variables training
```

```python
X_test #x-variable testing
```

```python
y_train #y-variable training
```

```python
y_test #y-variable testing
```

# **Model Building**

##Linear Regression

```python
from sklearn.linear_model import LinearRegression
```

```python
lr = LinearRegression()
```

### Training the model - the model is learning from X_train dataset and y_train correct answers(labels)

```python
lr.fit(X_train, y_train)
```

```python
#that's all, the model is built
```

### Applying the model to make a prediction

```python
y_lr_train_pred = lr.predict(X_train)
```

```python
y_lr_test_pred = lr.predict(X_test)
```

```python
y_lr_train_pred #these are predictions of 80% of the data
```

```python
y_lr_test_pred #these are the predictions of 20% of the data
```

### Evaluate model performance

```python
from sklearn.metrics import mean_squared_error, r2_score
```

```python
lr_train_mse = mean_squared_error(y_train, y_lr_train_pred)
lr_train_r2 = r2_score(y_train, y_lr_train_pred)
lr_test_mse = mean_squared_error(y_test, y_lr_test_pred)
lr_test_r2 = r2_score(y_test, y_lr_test_pred)
```

```python
print('LR MSE (Train): ', lr_train_mse)
print('LR R2 (Train): ', lr_train_r2)
print('LR MSE (Test): ', lr_test_mse)
print('LR R2 (Test): ', lr_test_r2)
```

```python
lr_results = pd.DataFrame(['Linear regression', lr_train_mse, lr_train_r2, lr_test_mse, lr_test_r2]).transpose()
lr_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
lr_results
```

## **Random Forest**

### Training the model

```python
from sklearn.ensemble import RandomForestRegressor
```

random forest has two versions - RandomForestRegressor and RandomForestClassifier
RandoForestRegressor - is a regression model (for quantitative data)
RandomForestClassifier - is a classification model (for categorical data)
Since logF(label) is quantitative - we are using a regression model

```python
rf = RandomForestRegressor(max_depth = 2, random_state = 100)
rf.fit(X_train, y_train)
```

### Applying the model to make a prediction

```python
y_rf_train_pred = rf.predict(X_train)
y_rf_test_pred = rf.predict(X_test)
```

```python
y_rf_train_pred#predicting y_train
```

```python
y_rf_test_pred#predicting y_test
```

### Evaluate model performance

```python
rf_train_mse = mean_squared_error(y_train, y_rf_train_pred)
rf_train_r2 = r2_score(y_train, y_rf_train_pred)
rf_test_mse = mean_squared_error(y_test, y_rf_test_pred)
rf_test_r2 = r2_score(y_test, y_rf_test_pred)
```

```python
print('RF MSE (Train): ', rf_train_mse)
print('RF R2 (Train): ', rf_train_r2)
print('RF MSE (Test): ', rf_test_mse)
print('RF R2 (Test): ', rf_test_r2)
```

```python
rf_results = pd.DataFrame(['Random Forest', rf_train_mse, rf_train_r2, rf_test_mse, rf_test_r2]).transpose()
rf_results.columns = ['Method', 'Training MSE', 'Training R2', 'Test MSE', 'Test R2']
```

```python
rf_results
```

Now we have 2 models to compare each-other with: LR and RF models

# Model Comparison

```python
df_models = pd.concat([lr_results, rf_results], axis = 0)
df_models
```

```python
df_models.reset_index(drop = True)
```

# Data Visualization of prediction results vs. Actual values

```python
import matplotlib.pyplot as plt
import numpy as np

plt.figure(figsize = (5,5))
plt.scatter(x = y_train, y = y_lr_train_pred, color = 'pink', alpha = 0.3)

z = np.polyfit(y_train, y_lr_train_pred, 1)
p = np.poly1d(z)
plt.plot(y_train, p(y_train), 'red') #best fit-line

plt.ylabel('Prediction logS')
plt.xlabel('Actual Logs')
plt.title('Linear Regression: Actual vs Predicted (Test Set)')
plt.grid(True)
```

If most points are close to the red line, prediction is good.
